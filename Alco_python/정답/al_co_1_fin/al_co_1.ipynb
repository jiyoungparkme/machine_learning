##### 기초 판다스 조작법_2
* 엑셀에서 흔히 사용하는 필터 기능을 구현
* 원하는 row와 column 을 따로 선별하는 방법을 다룹니다.
* 오름차순 / 내림차순하는 방법을 다룹니다.
* 기초 통계량을 알아봅니다.
* Transpose 에 대해 다룹니다.

```python
df['temp'] < 9
# 한 컬럼을 정하여 해당 컬럼 안에 있는 값들에 대해서
# 불리언 비교를 수행합니다.
```

```python
tf = df.loc[df['temp'] < 9]
# .loc 기능으로 불리언 인덱싱이라고 하는 문법을 적용합니다.
# 이 문법의 기능은 불리언 상 True 인 것을 선택하겠다는 것입니다.
# 전체 데이터 프레임인 df 에서 temp 컬럼 기준으로 9 보다 작은
# row들을 True 로 만드는 불리언을 만든 다음에
# 이를 기준으로 True 에 해당하는 row만 선택하는 것입니다.
print(tf.shape)
tf.head()
```

##### 컬럼에 대하여 수식 적용하기
* apply, lambda 이용합니다.

```python
lambda x : x / 10
```

```python
df['temp'].head()
```

```python
df['temp'].apply(lambda x : x / 10)
# .apply 이용하여 컬럼 내 하나하나 값들에 대해
# 수식을 적용할 수 있습니다.
# 여기에서 lambda 문법이 사용됩니다.
# lambda x : x / 10 은 이 함수에 들어오는 값 x 에 대하여 10으로 나누겠다는 것입니다.
# 따라서 .apply 진행할 경우 해당 컬럼에 있는 모든 값들을 하나하나씩 lambda 함수에 넣어서
# 각 값들을 10으로 나눈 것으로 변환하는 것입니다.
```

```python
df['temp'] / 10
# 아주 간단한 수식의 경우 .apply(lambda ~ ) 할 필요 없이
# 기초 사직연산(+, -, *, /) 가능합니다.
```
