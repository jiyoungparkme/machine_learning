### 기준 정했을 때 정도의 차이가 발생하는가?
* DepositType column -> 예치금/환불규정
* 취소시 손님이 넘어야 할 심리적 장벽(?)을 기준으로 삼으면 크고 작음을 나타낼 수 있음

```python
h1["DepositType"].unique()
```

```python
h1["DepositType"] = h1["DepositType"].apply(lambda x : x.strip())
h1["DepositType"].unique()
```

```python
mapper = {'No Deposit':0,'Refundable':1,'Non Refund':2}
h1["CustomerRisk"] = h1["DepositType"].map(mapper)
h1["CustomerRisk"].unique()
```

#### 연습문제 - Meal 컬럼 변환
##### 예약시 식사를 같이 결제한 정보입니다
* Undefinded 또는 SC인 경우 -> 식사 예약 X
* 식사 예약을 많이 할 수록 BB->HB->FB
* BB - Bed & Breakfast, HB - Half-Board, FB - Full-Board
* 따라서 Meal 컬럼은 호텔에 얼마나 길게 체류할 것인가를 나타냅니다.
* Meal 컬럼의 값을 0,1,2,3 으로 정도를 ㅡ나타내게 변환하여 Stay 컬럼을 만들어줍니다

```python
h1["Meal"] = h1["Meal"].apply(lambda x: x.strip())
mapper = {'SC': 0,'Undefined': 0,'BB': 1,'HB': 2,'FB': 3}
h1["Stay"] = h1["Meal"].map(mapper)
h1["Stay"].unique()
```

# 알고리즘에서 나오는 용어 정리
##### ensemble(앙상블)
* 최종 결과를 내기 위해서 작은 알고리즘들을 여러 개 이용하는 경우
* 숫자를 예측하는 경우에는 알고리즘들의 결과값을 평균
* 경우의 수를 예측하는 경우(분류)에는 알고리즘들의 결과를 투표

##### bagging(배깅) = bootstrap  + aggregation
* bootstrap? -> 데이터의 row를 추출해서 새로운 데이터 셋 만들기(복원추출)
* 복원추출? -> 뽑은 것을 빼놓지 않고 다시 넣어놓고 뽑음
* sklearn.ensemvle import RandomForestRegressor -> random forest 알고리즘ㅈ
* bagging을 이용하는 가장 대표적인 알고리즘
* 서로 조금씩 다른 데이터 셋을 bootstrap 방식으로 만들어내고
* 각 데이터 셋에 대해 의사결정 나무 알고리즘으로  따로따로 학습
* 그렇게 만들어진 여러 개의 의사결정나무 알고리즘의 결과를 합산(aggregate)하여 결과를 냄

##### boosting
* 오답노트와 유사한 접근법
* 학습 후 예측하면 잘 맞추는 문제(row)와 못 맞추는 문제를 알 수 있습니다
* 못 맞추는 문제(row)를 더 잘 맞출 수 있게하는 접근 방식
* 예_1) 하나 학습해서 예측한 다음 못 맞추는 row를 더 샘플링으로 뽑아서
* 새로운 데이터 셋 만들고 새로운 알고리즘 만들어서 학습
* 예_2) 한 알고리즘이 예측할 떄 맞춘 row와 못 맞춘 row가 있을텐데
* 못 맞춘 정도를 새로운 label로 삼아서 동일한 데이터에 대해 label만 바꿔서
* 새로운 알고리즘으로 학습 
* 통계적 머신러닝 계열에서 최고의 성능을 자랑(딥러닝 포함해서 성능이 좋은 경우 많음)
* XGBoost, lightgbm이 대표적인 예시
* 잘 못 쓰면 overfit(과적합, train good / test bad) 무조건 발생
* 예측력이 좋음 -> 해석력은 떨어짐

#### 분류 문제의 평가지표
* confusion matrix(혼동행렬)

```python
from sklearn.metrics import confusion_matrix as cm
cm_data = cm(for_plot['actual'], for_plot['predict'])
TN, FP, FN, TP = cm_data.ravel()
print(TN, FP, FN, TP)
```

* TF -> 맞춘 것(t로시작)이 True, 틀린 것이 False
* PN -> 예측 결과가 1이 Positive, 0이 Negative
* 위 4개는 TF에서 하나, PN에서 하나 뽑아서 만든 이름
* TPR, TNR, PPV는 시험 준비할 때 외워야 됨 -> 위키피디ㅣ아 Confusion matrix 수식 참고

##### 화재경보장치  - 1이 알람 울림, 0이 알람 안울림
* FP = False Positive -> False Alarm (1종 오류)
* FN = False Negative -> Miss (2종 오류 더 심각한 오류)

##### 환자 검진에 사용시 - 1이 양성(걸렸다), 0이 음성(안 걸렸다)
* FP = False Positive -> 검사 결과 양성이었으나 사실은 정상(1종 오류)
* FN = False Negative -> 검사 결과 음성이었으나 사실은 병에 걸림(2종 오류 더 심각한 오류)

##### 금융 사기 탐지에 사용시  - 1이 사기라고 판단, 0이 사기 아니라고 판단
* FP = False Positive -> 사기인줄 알았으나 사기가 아니었음 (1종 오류)
* FN = False Negative -> 사기 아닌 줄 알았는데 사실은 사기 당한거였음 (2종 오류 더 심각한 오류)

##### 스팸 메일 분류에 사용시  - 1이 스팸이라고 판단, 0이 스팸 아니라고 판단 - FP가 더 중요
* FP = False Positive -> 스팸인줄 알았으나 스팸이 아니었음
* FN = False Negative -> 스팸 아닌 줄 알았는데 사실은 스팸이었음 

##### 어뷰징 유저 판단에 사용시  - 1이 어뷰징이라고 판단, 0이 어뷰징 아니라고 판단  - FP가 더 중요
* FP = False Positive -> 어뷰징이라 판단했으나 아니었음 
* FN = False Negative -> 어뷰징이 아니라 판단했는데 어뷰징이었음 

* FN을 0으로 만들면 2종오류는 없앨 수 있지만 FP가 올라감 (생사람 잡는 격)\
* 반대로 FP 줄이면 FN 증가

##### Trade-off 관계
* 하나를 얻으면 하나를 잃는 상황
* 모든 것을 얻을 수 없다는 의미
* 예) 노트북 화면 크고 성능 좋은데 가벼웠으면 좋겠음, 그리고 100만원 이하 -> 없음ㅋㅋㅋㅋ


# accuracy, precision, recall, fall-out
* accuracy = 전체 중에서 맞춘 비율
* precision(sensitivity) -> 예측 모델이 1(Positive)로 예측한 것 중에서 맞춘 것 비율 TP / (TP + FP)
* recall(specifity) -> 실제값이 1인 것 중에서 맞춘 것 비율 TP / (TP + FN)
* precision이 올라가면 recall 떨어짐(FP, FN 관계 때문임)
* f1-score = 2*precision * recall / (precision + recall)
* fall-out -> 실제값이 0인 것 중 틀린 것 비율 FP / (FP + TN)

```python
acc = (TN + TP) / (TP + TN + FP + FN)
acc
```

```python
prec = TP / (TP + FP) # F - 틀림, P - 1이라고 예측 
prec # 알고리즘이 1 이라고 했으면 얼마나 믿을 수 있나?
```

# ROC Curve? 어떻게 그려지는 것인가?

* 확률과 실제 정답을 이용하여 recall 과 fall-out 점수를 좌표로 사용
* 판단 기준을 0% -> 100% 로 바꿔가며 그때마다 confusion-matrix 얻고
* 각각 경우마다 recall과 fall-out 점수를 구합니다.
* 예) 0, 10, 20, 30, ... 100% 순으로 판단기준을 나누어보면
* confusion matrix가 11개 나옵니다.
* 11가지 경우의 수에 대해 recall, fall-out 구하고
* recall을 y좌표, fall-out을 x 좌표로 그리면 ROC curve 입니다.

```python
r2["predict"] = 0
r2.loc[r2["pos"]>0.91, "predict"]=1 # 부등호 값을 0~1 사이로 바꿔보세요
# 판단 기준을 0 -> 1로 점차 올려가면서 confusion matrix를 그릴 수 있겠습니다
cm_data = cm(r2["actual"],r2["predict"])
cm_data
```

```python
all_data = []
for i in range(100):
    r2["predict"] = 0
    r2.loc[r2["pos"]> i * 0.01, "predict"]=1
    # 반복마다 판단 기준을 조금씩 큰 값을 넣어줍니다
    cm_data = cm(r2["actual"],r2["predict"])
    TN, FP, FN, TP = cm_data.ravel()
    # 판단 기준이 달라질 때마다 confusion matrix를 새로 그릴 수 있음
    prec = TP / (TP + FP)
    recall = TP / (TP + FN)
    fall_out = FP / (FP + TN)
    # 새로 뽑은 confusion matrix로 평가지표를 뽑아냅니다.
    all_data.append({"prec" : prec, "recall" : recall, "fall_out" : fall_out})
```

```python
roc = pd.DataFrame(all_data)
roc
```

```python
def scorer(model):
    model.fit(X_train, y_train)
    for_plot= pd.DataFrame()
    for_plot["predict"] = model.predict(X_test)
    for_plot["actual"] = y_test
    cm_data = cm(for_plot["actual"],for_plot["predict"])
    TN, FP, FN, TP = cm_data.ravel()
    recall = TP / (TP+FN)
    return recall
```

```python
from sklearn.ensemble import RandomForestClassifier as rf
model =gb(learning_rate = 0.01, n_estimators=100, 
          max_depth=15, max_features=0.9)
scorer(model)
```

### 어떻게 접근할까?
##### 1. grid search -> 격자 모양으로 찾기 (데이터 손실 적음 but 시간 오래 걸림)
##### 2. random search -> 무작위로 찾기(나름의 체계가 있음)
##### 3. optimizer(최신 머신러닝에서 할거임) - 검색 키워드 : bayesian optimizer

```python
n_estimators = [3,4,5,6]
learning_rate = [0.01,0.05,0.1]
max_features = [0.1,0.3,0.5,0.7,0.9]
max_depth = [3,5,7,9,12]
all_scores = []
for a in n_estimators:
    for b in learning_rate:
        for c in max_features:
            for d in max_depth:
                model =gb(learning_rate = b, n_estimators=a, 
                          max_depth=d, max_features=c)
                # a, b, c, d 에는 각 하이퍼파라미터들의 각각 경우의 수에 해당하는 
                # 값들이 들어있습니다. 이를 이용해
                # 옵션을 설정 후 model을 만들어 줍니다.
                recall = scorer(model)
                all_scores.append({"a": a, "b":b, "c":c, "d":d, "score":recall})
                
score_df = pd.DataFrame(all_scores)
```

```python
score_df.sort_values(by="score", ascending = False).head()
```

```python
import time
from random import randint
all_scores = []
t0 = time.time()
while True:
    if time.time() - t0 > 30 : break
    a = randint(7, 10)
    b = randint(472, 838) / 1000
    c = randint(284, 828) / 1000
    d = randint(9, 10)
    model = gb(learning_rate=b, n_estimators=a, 
               max_depth=d, max_features=c)
    # a, b, c, d 에는 각 하이퍼파라미터들의 각각 경우의 수에 해당하는
    # 값들이 들어있습니다. 이를 이용해
    # 옵션을 설정 후 model 을 만들어줍니다.
    recall = scorer(model)
    all_scores.append({'a':a, 'b':b, 'c':c, 'd':d, 'score':recall})
```
